```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert results dict to DataFrame
results_df = pd.DataFrame({
    model: {
        'R2': metrics['R2'],
        'RMSE': metrics['RMSE'],
        'MAE': metrics['MAE']
    }
    for model, metrics in results.items()
}).T  # Transpose so models are rows

# Rank models (R² higher better, RMSE & MAE lower better)
results_df['R2_rank'] = results_df['R2'].rank(ascending=False)
results_df['RMSE_rank'] = results_df['RMSE'].rank(ascending=True)
results_df['MAE_rank'] = results_df['MAE'].rank(ascending=True)

# Calculate total rank
results_df['Total_rank'] = results_df[['R2_rank', 'RMSE_rank', 'MAE_rank']].sum(axis=1)

# Find best model
best_model = results_df['Total_rank'].idxmin()

print("Performance Table:")
print(results_df[['R2', 'RMSE', 'MAE', 'Total_rank']])
print(f"\n🏆 Best performer overall: **{best_model}**")

# Plot comparison
results_df[['R2', 'RMSE', 'MAE']].plot(kind='bar', figsize=(8,6), subplots=True, layout=(1,3), sharex=True)
plt.suptitle("Model Performance Comparison", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
def plot_predictions(y_true, y_pred, model_name, r2, rmse, mae):
    plt.figure(figsize=(6,6))

    # Scatter plot
    plt.scatter(y_true, y_pred, alpha=0.6, edgecolor='k', s=50)

    # Reference line (perfect prediction)
    min_val, max_val = y_true.min(), y_true.max()
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', lw=2)

    # Labels and title with metrics
    plt.xlabel('Actual total_updrs')
    plt.ylabel('Predicted total_updrs')
    plt.title(f"{model_name}\nR²={r2:.3f} | RMSE={rmse:.3f} | MAE={mae:.3f}")

    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Loop through models and plot
for model_name, metrics in results.items():
    plot_predictions(
        y_test,
        metrics['y_pred'],
        model_name,
        metrics['R2'],
        metrics['RMSE'],
        metrics['MAE']
    )

```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# Base model
rf = RandomForestRegressor(random_state=42)

# Parameter search space
param_dist_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None],
    'bootstrap': [True, False]
}

# Randomized Search
rand_rf = RandomizedSearchCV(
    rf,
    param_distributions=param_dist_rf,
    n_iter=20,        # Try 20 random combinations
    cv=3,             # 3-fold cross-validation
    scoring='r2',
    n_jobs=-1,
    verbose=2,
    random_state=42
)

# Fit
rand_rf.fit(X_train, y_train)

# Best params
print("Best Parameters (Random Forest):", rand_rf.best_params_)
print("Best CV R²:", rand_rf.best_score_)

# Test set evaluation
best_rf = rand_rf.best_estimator_
y_pred_rf_best = best_rf.predict(X_test)

r2_rf_best = r2_score(y_test, y_pred_rf_best)
rmse_rf_best = np.sqrt(mean_squared_error(y_test, y_pred_rf_best))
mae_rf_best = mean_absolute_error(y_test, y_pred_rf_best)

print(f"Tuned Random Forest -> R²: {r2_rf_best:.6f}, RMSE: {rmse_rf_best:.6f}, MAE: {mae_rf_best:.6f}")
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import numpy as np

# Pipeline
mlp_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('mlp', MLPRegressor(random_state=42, max_iter=500))
])

# Smaller parameter space
param_dist_mlp = {
    'mlp__hidden_layer_sizes': [(50,), (100,), (50,50)],
    'mlp__activation': ['relu', 'tanh'],
    'mlp__solver': ['adam'],
    'mlp__alpha': [0.0001, 0.001, 0.01],
    'mlp__learning_rate': ['constant', 'adaptive']
}

# Randomized search
rand_mlp = RandomizedSearchCV(
    mlp_pipe,
    param_distributions=param_dist_mlp,
    n_iter=10,  # Only try 10 random combinations
    cv=3,
    scoring='r2',
    n_jobs=-1,
    verbose=2,
    random_state=42
)

rand_mlp.fit(X_train, y_train)

print("Best Parameters:", rand_mlp.best_params_)
print("Best CV R²:", rand_mlp.best_score_)

# Test set evaluation
best_mlp = rand_mlp.best_estimator_
y_pred_mlp_best = best_mlp.predict(X_test)

r2_mlp_best = r2_score(y_test, y_pred_mlp_best)
rmse_mlp_best = np.sqrt(mean_squared_error(y_test, y_pred_mlp_best))
mae_mlp_best = mean_absolute_error(y_test, y_pred_mlp_best)

print(f"Tuned MLP -> R²: {r2_mlp_best:.6f}, RMSE: {rmse_mlp_best:.6f}, MAE: {mae_mlp_best:.6f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error

# ====== BEFORE tuning results ======
metrics_before = {
    "Decision Tree": [
        r2_dt,  # R²
        np.sqrt(mean_squared_error(y_test, y_pred_dt)),  # RMSE
        mean_absolute_error(y_test, y_pred_dt)  # MAE
    ],
    "MLP": [
        r2_mlp,
        np.sqrt(mean_squared_error(y_test, y_pred_mlp)),
        mean_absolute_error(y_test, y_pred_mlp)
    ],
    "Random Forest": [
        r2_rf,
        np.sqrt(mean_squared_error(y_test, y_pred_rf)),
        mean_absolute_error(y_test, y_pred_rf)
    ]
}

# ====== AFTER tuning results ======
metrics_after = {
    "Decision Tree": [r2_dt_best, rmse_dt_best, mae_dt_best],
    "MLP": [r2_mlp_best, rmse_mlp_best, mae_mlp_best],
    "Random Forest": [r2_rf_best, rmse_rf_best, mae_rf_best]
}

# ====== Plotting ======
metrics_names = ["R²", "RMSE", "MAE"]
models = list(metrics_before.keys())
x = np.arange(len(models))
width = 0.35

for i, metric in enumerate(metrics_names):
    before_vals = [metrics_before[m][i] for m in models]
    after_vals = [metrics_after[m][i] for m in models]

    fig, ax = plt.subplots()
    ax.bar(x - width/2, before_vals, width, label='Before Tuning')
    ax.bar(x + width/2, after_vals, width, label='After Tuning')

    ax.set_ylabel(metric)
    ax.set_title(f'{metric} Before vs After Tuning')
    ax.set_xticks(x)
    ax.set_xticklabels(models)
    ax.legend()
    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Annotate values on bars
    for j, v in enumerate(before_vals):
        ax.text(j - width/2, v + (0.01 if metric != "RMSE" else 0.05), f"{v:.3f}", ha='center')
    for j, v in enumerate(after_vals):
        ax.text(j + width/2, v + (0.01 if metric != "RMSE" else 0.05), f"{v:.3f}", ha='center')

    plt.show()
```

```python
import matplotlib.pyplot as plt

def plot_actual_vs_pred(ax, y_true, y_pred, title):
    ax.scatter(y_true, y_pred, alpha=0.5)
    ax.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--')
    ax.set_xlabel('Actual total_updrs')
    ax.set_ylabel('Predicted total_updrs')
    ax.set_title(title)
    ax.grid(True, linestyle='--', alpha=0.6)

# ===== Data to plot (Before & After tuning for 3 models) =====
models_data = [
    ("Decision Tree", y_pred_dt, y_pred_dt_best),
    ("MLP", y_pred_mlp, y_pred_mlp_best),
    ("Random Forest", y_pred_rf, y_pred_rf_best)
]

# ===== Create subplots dynamically =====
fig, axes = plt.subplots(len(models_data), 2, figsize=(12, 5 * len(models_data)))

for i, (name, pred_before, pred_after) in enumerate(models_data):
    plot_actual_vs_pred(axes[i, 0], y_test, pred_before, f"{name} - Before Tuning")
    plot_actual_vs_pred(axes[i, 1], y_test, pred_after, f"{name} - After Tuning")

plt.tight_layout()
plt.show()


```
